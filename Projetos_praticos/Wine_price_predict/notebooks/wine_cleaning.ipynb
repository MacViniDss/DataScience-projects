```python
import pandas as pd
import numpy as np

df= pd.read_csv('winemag-data-130k-v2.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df.columns
```

```python
df.drop(['Unnamed: 0','region_2'], axis= 1, inplace= True)
df = df.drop_duplicates()
```

```python
df.columns
```

```python
df.isna().sum()
```

```python
countryna = df.loc[df.country.isna()]
wineryna = list(countryna.winery.unique())
wineryna
```

```python
df['country'] = df.groupby('winery')['country'].transform(lambda x: x.fillna(x.mode().iloc[0]) if not x.mode().empty else x)
df['region_1'] = df.groupby('winery')['region_1'].transform(lambda x: x.fillna(x.mode().iloc[0]) if not x.mode().empty else x)
df['province'] = df.groupby('winery')['province'].transform(lambda x: x.fillna(x.mode().iloc[0]) if not x.mode().empty else x)
```

```python
df['country'].isna().sum(),df['region_1'].isna().sum(),df['province'].isna().sum() 
```

```python
df.loc[df.country.isna()]
```

```python
df.loc[df.country.isna()].winery.value_counts()
df.to_csv("wine_clean.csv", index=False)
```

```python
df['region_1'] = df.groupby(['country'])['region_1'].transform(lambda x: x.fillna(x.mode().iloc[0]) if not x.mode().empty else x)
```

```python
regionna = df.loc[df.region_1.isna()]
regionna.country.unique()
```

```python
linhas_remover = df.loc[df.country.isna() | df.region_1.isna()].index
df.drop(index = linhas_remover, inplace= True)
```

```python
df.isna().sum()
```

```python
varityna = df.loc[df.variety.isna()].index
df.drop(index = varityna, inplace= True)
df.isna().sum()
```

```python
pricena = df.loc[df.price.isna()]
pricena.title.nunique()
```

```python
df = df[(df['price'] < df['price'].quantile(0.99)) &
        (df['points'] < df['points'].quantile(0.99))]

df['price'] = df.groupby(['country', 'variety', 'points'])['price'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isna().sum()
```

```python
categorical_columns = list(df.dtypes[df.dtypes == 'object'].index)
# Aqui faço uma limpeza nas str
for c in categorical_columns:
    df[c] = df[c].str.lower().str.replace(' ', '_')
    df[c] = df[c].str.lower().str.replace('-', '_')
df.info()
```
